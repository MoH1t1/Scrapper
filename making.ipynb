```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
import logging
logging.basicConfig(filename="scrapper.log" , level=logging.INFO)
```

```python
flipkart_url = "https://www.flipkart.com/search?q=" + "iphone12promax"
print(flipkart_url)        # through this we can go to that page
```

```python
# If we want to go to that page without using the URL, we can use the urllib lib.
urlclient = uReq(flipkart_url)
flipkart_page = urlclient.read()                   # Returns a data dump (view page source data)
print(flipkart_page)
```

```python
# Now we will use beatify to convert the data dump in human readable form
flipkart_html = bs(flipkart_page, "html.parser")
print(flipkart_html)
```

```python
# Now we have to retrieve the link of each product (href link) to be able to scrape the product details
bigbox = flipkart_html.findAll("div", {"class":"cPHDOP col-12-12"} )
```

```python
len(bigbox)
```

```python
# Retrieving one box
bigbox[3]
```

```python
# Retrieving data(link) from one box
productlink = "https://www.flipkart.com" + bigbox[3].div.div.div.a['href']
print(productlink)
```

```python
product_req = requests.get(productlink)            # Extracts data from that product link
product_html = bs(product_req.text, "html.parser") # beautifies it
```

```python
comment_box = product_html.find_all("div",{"class": "RcXBOT"})  # Extracts review section
print(comment_box)
```

```python
comment_box[0].div.div.find_all('p',{"class":"_2NsDsF AwS1CA"})[0].text
```

```python
for i in comment_box:
    print(i.div.div.find_all('p',{"class":"_2NsDsF AwS1CA"})[0].text)                     # Extarcting all name
```

```python
for i in comment_box:
    print(i.div.div.div.div.text)                # Retrieving all rating
```

```python
for i in comment_box:
    print(i.div.div.div.p.text)               # Extract comment header
```

```python
for i in comment_box:
    print(i.div.div.find_all('div',{"class":""})[0].text)                      # Extarct Comments
```
